```python
# Interface conversacional 
while True:

    question = input(
        "\nCyberSentinel > "
    )


    if question.lower() in [
        "sair",
        "exit",
        "quit"
    ]:
        break


    response = agent.ask(
        question
    )


    print(
        "\nResposta:\n"
    )

    print(response)
```
